```python
import torch
import torchvision
import torchvision.transforms as transforms
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, random_split, Subset
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, classification_report
import seaborn as sns
import numpy as np
import time
import random
import torch.nn.functional as F

def set_seed(seed_value=42):
    """Sets seed for reproducibility."""
    random.seed(seed_value)
    np.random.seed(seed_value)
    torch.manual_seed(seed_value)
    if torch.cuda.is_available():
        torch.cuda.manual_seed(seed_value)
        torch.cuda.manual_seed_all(seed_value)
        torch.backends.cudnn.deterministic = True
        torch.backends.cudnn.benchmark = False

set_seed(42)

device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
print(f"Using device: {device}")

classes = ('plane', 'car', 'bird', 'cat', 'deer',
           'dog', 'frog', 'horse', 'ship', 'truck')

IMAGENET_MEAN = [0.485, 0.456, 0.406]
IMAGENET_STD = [0.229, 0.224, 0.225]

transform_train_resnet = transforms.Compose([
    transforms.RandomCrop(32, padding=4),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(degrees=15),
    transforms.ColorJitter(brightness=0.2, contrast=0.2, saturation=0.2, hue=0.1),
    transforms.ToTensor(),
    transforms.Normalize(IMAGENET_MEAN, IMAGENET_STD),
    transforms.RandomErasing(p=0.5, scale=(0.02, 0.2), ratio=(0.3, 3.3), value='random')
])

transform_train_basic = transforms.Compose([
    transforms.RandomHorizontalFlip(),
    transforms.RandomCrop(32, padding=4),
    transforms.ToTensor(),
    transforms.Normalize(IMAGENET_MEAN, IMAGENET_STD)
])

transform_test = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize(IMAGENET_MEAN, IMAGENET_STD)
])

def imshow(img_tensor, title=None, ax=None):
    """Displays a Tensor image after unnormalization."""
    img_tensor = img_tensor.cpu().clone()
    mean_t = torch.tensor(IMAGENET_MEAN).view(3, 1, 1)
    std_t = torch.tensor(IMAGENET_STD).view(3, 1, 1)
    img_tensor = img_tensor * std_t + mean_t
    img_tensor = torch.clamp(img_tensor, 0, 1)
    npimg = img_tensor.numpy()

    if ax is None:
        plt.imshow(np.transpose(npimg, (1, 2, 0)))
        if title: plt.title(title)
    else:
        ax.imshow(np.transpose(npimg, (1, 2, 0)))
        if title: ax.set_title(title, fontsize=8)
        ax.axis('off')

full_train_dataset_orig = torchvision.datasets.CIFAR10(root='./data', train=True, download=True)
test_dataset_transformed = torchvision.datasets.CIFAR10(root='./data', train=False, download=True, transform=transform_test)

orig_dataset_size = len(full_train_dataset_orig)
train_size = int(0.8 * orig_dataset_size)
val_size = orig_dataset_size - train_size

data_split_generator = torch.Generator().manual_seed(42)
train_subset_orig, val_subset_orig = random_split(
    full_train_dataset_orig,
    [train_size, val_size],
    generator=data_split_generator
)

class DatasetTransformer(torch.utils.data.Dataset):
    def __init__(self, base_dataset, transform_fn):
        self.base_dataset = base_dataset
        self.transform_fn = transform_fn

    def __getitem__(self, index):
        img, target = self.base_dataset[index]
        if self.transform_fn:
            img = self.transform_fn(img)
        return img, target

    def __len__(self):
        return len(self.base_dataset)

train_dataset_basic = DatasetTransformer(train_subset_orig, transform_train_basic)
val_dataset_basic = DatasetTransformer(val_subset_orig, transform_test)

train_dataset_resnet = DatasetTransformer(train_subset_orig, transform_train_resnet)
val_dataset_resnet = DatasetTransformer(val_subset_orig, transform_test)

BATCH_SIZE = 128
NUM_WORKERS = 2

train_loader_basic = DataLoader(train_dataset_basic, batch_size=BATCH_SIZE, shuffle=True, num_workers=NUM_WORKERS, pin_memory=True)
val_loader_basic = DataLoader(val_dataset_basic, batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS, pin_memory=True)

train_loader_resnet = DataLoader(train_dataset_resnet, batch_size=BATCH_SIZE, shuffle=True, num_workers=NUM_WORKERS, pin_memory=True)
val_loader_resnet = DataLoader(val_dataset_resnet, batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS, pin_memory=True)

test_loader = DataLoader(test_dataset_transformed, batch_size=BATCH_SIZE, shuffle=False, num_workers=NUM_WORKERS, pin_memory=True)


class MLP(nn.Module):
    def __init__(self, input_dim=3*32*32, num_classes=10):
        super(MLP, self).__init__()
        self.flatten = nn.Flatten()
        h1_size, h2_size, h3_size = 512, 256, 128
        self.fc_layers = nn.Sequential(
            nn.Linear(input_dim, h1_size), nn.ReLU(), nn.Dropout(0.3),
            nn.Linear(h1_size, h2_size), nn.ReLU(), nn.Dropout(0.3),
            nn.Linear(h2_size, h3_size), nn.ReLU(),
            nn.Linear(h3_size, num_classes)
        )
    def forward(self, x):
        x = self.flatten(x)
        return self.fc_layers(x)

class CNN(nn.Module):
    def __init__(self, num_classes=10):
        super(CNN, self).__init__()
        self.conv1 = nn.Sequential(
            nn.Conv2d(3, 64, 3, padding=1), nn.BatchNorm2d(64), nn.ReLU(),
            nn.Conv2d(64, 64, 3, padding=1), nn.BatchNorm2d(64), nn.ReLU(),
            nn.MaxPool2d(2, 2)) # 32x32 -> 16x16
        self.conv2 = nn.Sequential(
            nn.Conv2d(64, 128, 3, padding=1), nn.BatchNorm2d(128), nn.ReLU(),
            nn.Conv2d(128, 128, 3, padding=1), nn.BatchNorm2d(128), nn.ReLU(),
            nn.MaxPool2d(2, 2)) # 16x16 -> 8x8
        self.conv3 = nn.Sequential(
            nn.Conv2d(128, 256, 3, padding=1), nn.BatchNorm2d(256), nn.ReLU(),
            nn.Conv2d(256, 256, 3, padding=1), nn.BatchNorm2d(256), nn.ReLU(),
            nn.MaxPool2d(2, 2)) # 8x8 -> 4x4
        self.flatten = nn.Flatten()
        fc_input_dim = 256 * 4 * 4
        self.fc_layers = nn.Sequential(
            nn.Linear(fc_input_dim, 512), nn.ReLU(), nn.Dropout(0.5),
            nn.Linear(512, num_classes))

    def forward(self, x):
        x = self.conv1(x)
        x = self.conv2(x)
        x = self.conv3(x)
        x = self.flatten(x)
        return self.fc_layers(x)

def get_resnet18(num_classes=10):
    model = torchvision.models.resnet18(weights=torchvision.models.ResNet18_Weights.IMAGENET1K_V1)
    model.fc = nn.Linear(model.fc.in_features, num_classes)
    return model

def train_model(model, model_name, train_dl, val_dl, criterion, optimizer, scheduler, epochs=20, one_cycle_lr=False):
    start_time = time.time()
    train_losses, val_losses, train_accs, val_accs = [], [], [], []
    best_model_path = f"{model_name}_cifar10_best.pth"
    best_val_acc = 0.0

    print(f"Starting training: {model_name} for {epochs} epochs...")
    for epoch in range(epochs):
        epoch_start_time = time.time()
        model.train()
        current_train_loss, correct_train, total_train = 0.0, 0, 0
        for i, (inputs, labels) in enumerate(train_dl):
            inputs, labels = inputs.to(device), labels.to(device)
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()
            if one_cycle_lr and scheduler: scheduler.step()

            current_train_loss += loss.item() * inputs.size(0)
            _, predicted = torch.max(outputs.data, 1)
            total_train += labels.size(0)
            correct_train += (predicted == labels).sum().item()

            if (i + 1) % 100 == 0:
                lr = optimizer.param_groups[0]["lr"]
                print(f'{model_name} - E[{epoch+1}/{epochs}], B[{i+1}/{len(train_dl)}], Loss: {loss.item():.4f}, LR: {lr:.6f}')

        avg_train_loss = current_train_loss / total_train if total_train > 0 else 0
        avg_train_acc = correct_train / total_train if total_train > 0 else 0
        train_losses.append(avg_train_loss); train_accs.append(avg_train_acc)

        model.eval()
        current_val_loss, correct_val, total_val = 0.0, 0, 0
        with torch.no_grad():
            for inputs, labels in val_dl:
                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model(inputs)
                loss = criterion(outputs, labels)
                current_val_loss += loss.item() * inputs.size(0)
                _, predicted = torch.max(outputs.data, 1)
                total_val += labels.size(0)
                correct_val += (predicted == labels).sum().item()

        avg_val_loss = current_val_loss / total_val if total_val > 0 else 0
        avg_val_acc = correct_val / total_val if total_val > 0 else 0
        val_losses.append(avg_val_loss); val_accs.append(avg_val_acc)

        epoch_duration = time.time() - epoch_start_time
        print(f'{model_name} - E{epoch+1} ({epoch_duration:.2f}s). Train L:{avg_train_loss:.4f},Acc:{avg_train_acc:.4f} | Val L:{avg_val_loss:.4f},Acc:{avg_val_acc:.4f}')

        if scheduler and not one_cycle_lr:
            if isinstance(scheduler, torch.optim.lr_scheduler.ReduceLROnPlateau): scheduler.step(avg_val_loss)
            else: scheduler.step()
        if avg_val_acc > best_val_acc:
            best_val_acc = avg_val_acc
            torch.save(model.state_dict(), best_model_path)
            print(f"Saved best {model_name} (Val Acc: {best_val_acc:.4f})")

    total_train_time = time.time() - start_time
    print(f'Training {model_name} finished in {total_train_time//60:.0f}m {total_train_time%60:.0f}s.')
    return train_losses, val_losses, train_accs, val_accs

def test_model(model, model_name, test_dl, criterion, class_names_list):
    best_model_path = f"{model_name}_cifar10_best.pth"
    try:
        model.load_state_dict(torch.load(best_model_path, map_location=device))
        print(f"\nLoaded best {model_name} from '{best_model_path}' for testing.")
    except FileNotFoundError:
        print(f"\nSaved model '{best_model_path}' not found. Using current model state.")
    except Exception as e:
        print(f"\nError loading model '{best_model_path}': {e}. Using current model state.")

    model.eval()
    current_test_loss, correct_test, total_test = 0.0, 0, 0
    all_preds, all_labels = [], []
    with torch.no_grad():
        for inputs, labels in test_dl:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            current_test_loss += loss.item() * inputs.size(0)
            _, predicted = torch.max(outputs.data, 1)
            total_test += labels.size(0)
            correct_test += (predicted == labels).sum().item()
            all_preds.extend(predicted.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())

    avg_test_loss = current_test_loss / total_test if total_test > 0 else 0
    avg_test_acc = correct_test / total_test if total_test > 0 else 0
    print(f'Test results for {model_name}: Loss: {avg_test_loss:.4f}, Accuracy: {avg_test_acc:.4f}')
    print("\nDetailed Classification Report:")
    print(classification_report(all_labels, all_preds, target_names=class_names_list, zero_division=0))
    return all_labels, all_preds, avg_test_acc

def plot_learning_curves(train_loss_hist, val_loss_hist, train_acc_hist, val_acc_hist, model_name_str):
    if not train_loss_hist:
        print(f"No learning curve data for {model_name_str}.")
        return
    epochs_range = range(1, len(train_loss_hist) + 1)
    plt.figure(figsize=(12, 5))
    plt.subplot(1, 2, 1)
    plt.plot(epochs_range, train_loss_hist, 'bo-', label='Train Loss')
    plt.plot(epochs_range, val_loss_hist, 'ro-', label='Val Loss')
    plt.title(f'{model_name_str} - Loss'); plt.xlabel('Epochs'); plt.ylabel('Loss'); plt.legend(); plt.grid(True)
    plt.subplot(1, 2, 2)
    plt.plot(epochs_range, train_acc_hist, 'bo-', label='Train Acc')
    plt.plot(epochs_range, val_acc_hist, 'ro-', label='Val Acc')
    plt.title(f'{model_name_str} - Accuracy'); plt.xlabel('Epochs'); plt.ylabel('Accuracy'); plt.legend(); plt.grid(True)
    plt.suptitle(f'Learning Curves: {model_name_str}'); plt.tight_layout(rect=[0, 0, 1, 0.96]); plt.show()

def plot_confusion_matrix(true_lbls, pred_lbls, class_names_list, model_name_str):
    if not true_lbls or not pred_lbls:
        print(f"No confusion matrix data for {model_name_str}.")
        return
    cm = confusion_matrix(true_lbls, pred_lbls)
    plt.figure(figsize=(10, 8))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names_list, yticklabels=class_names_list)
    plt.xlabel('Predicted Label'); plt.ylabel('True Label'); plt.title(f'Confusion Matrix: {model_name_str}'); plt.show()

def plot_probabilities_bar(probs_arr, true_label_idx_val, class_names_list, ax_obj):
    y_pos = np.arange(len(class_names_list))
    bars = ax_obj.barh(y_pos, probs_arr, align='center', height=0.7)
    ax_obj.set_yticks(y_pos)
    ax_obj.set_yticklabels(class_names_list, fontsize=7)
    ax_obj.invert_yaxis()
    ax_obj.set_xlabel('Probability', fontsize=8)
    if true_label_idx_val is not None and 0 <= true_label_idx_val < len(bars):
        bars[true_label_idx_val].set_color('green')
    ax_obj.tick_params(axis='x', labelsize=7)
    ax_obj.set_xlim(0, 1)

def visualize_predictions_with_probs(model_dict_param, data_loader_param, num_imgs=5, class_names_list=None):
    if not class_names_list: class_names_list = tuple(str(i) for i in range(10))
    try:
        img_iter = iter(data_loader_param)
        imgs, lbls = next(img_iter)
    except StopIteration:
        print("Cannot fetch data from DataLoader (possibly empty).")
        return

    imgs_to_show = imgs[:num_imgs].to(device)
    lbls_to_show = lbls[:num_imgs]
    active_models_count = sum(1 for m in model_dict_param.values() if m is not None)
    if active_models_count == 0:
        print("No trained models to visualize.")
        return

    fig_viz, axes_viz = plt.subplots(num_imgs, 1 + active_models_count,
                                 figsize=(3.5 * (1 + active_models_count), 3 * num_imgs + 0.5),
                                 squeeze=False)
    fig_viz.suptitle("Sample Predictions and Probability Distributions", fontsize=16, y=1.0)

    for i in range(num_imgs):
        ax_img_viz = axes_viz[i, 0]
        true_lbl_idx = lbls_to_show[i].item()
        true_lbl_name = class_names_list[true_lbl_idx]
        imshow(imgs_to_show[i].cpu(), title=f"True: {true_lbl_name}", ax=ax_img_viz)

        plot_col = 1
        for model_disp_name, model_obj_viz in model_dict_param.items():
            if model_obj_viz is None: continue
            ax_prob_viz = axes_viz[i, plot_col]
            model_obj_viz.eval().to(device)
            with torch.no_grad():
                output_viz = model_obj_viz(imgs_to_show[i].unsqueeze(0))
                probs_viz = F.softmax(output_viz, dim=1).squeeze().cpu().numpy()
            pred_idx = np.argmax(probs_viz)
            pred_lbl_name = class_names_list[pred_idx]
            status_str = "(Correct)" if pred_lbl_name == true_lbl_name else "(Incorrect)"
            ax_prob_viz.set_title(f"{model_disp_name}: {pred_lbl_name} {status_str}", fontsize=9)
            plot_probabilities_bar(probs_viz, true_lbl_idx, class_names_list, ax_prob_viz)
            plot_col += 1
    plt.tight_layout(rect=[0, 0, 1, 0.97]); plt.show()

EPOCHS_BASIC = 30
EPOCHS_RESNET = 20

LR_BASIC = 0.001
LR_RESNET_MAX = 0.01
WD_RESNET = 1e-4

common_criterion = nn.CrossEntropyLoss()
model_results = {}

mlp_model = MLP().to(device)
cnn_model = CNN().to(device)
resnet_model = get_resnet18().to(device)

print("\n Training MLP ")
mlp_optimizer = optim.Adam(mlp_model.parameters(), lr=LR_BASIC)
mlp_scheduler = optim.lr_scheduler.ReduceLROnPlateau(mlp_optimizer, 'min', patience=5, factor=0.2, verbose=True)
mlp_train_loss, mlp_val_loss, mlp_train_acc, mlp_val_acc = train_model(
    mlp_model, "MLP", train_loader_basic, val_loader_basic,
    common_criterion, mlp_optimizer, mlp_scheduler, epochs=EPOCHS_BASIC)
mlp_true_labels, mlp_preds, mlp_test_acc = test_model(
    mlp_model, "MLP", test_loader, common_criterion, classes)
model_results["MLP"] = {
    "train_loss_hist": mlp_train_loss, "val_loss_hist": mlp_val_loss, "train_acc_hist": mlp_train_acc, "val_acc_hist": mlp_val_acc,
    "test_labels": mlp_true_labels, "test_preds": mlp_preds, "test_acc": mlp_test_acc, "model_obj": mlp_model}
if mlp_train_loss: plot_learning_curves(mlp_train_loss, mlp_val_loss, mlp_train_acc, mlp_val_acc, "MLP")
if mlp_true_labels: plot_confusion_matrix(mlp_true_labels, mlp_preds, classes, "MLP")

print("\n Training CNN ")
cnn_optimizer = optim.AdamW(cnn_model.parameters(), lr=LR_BASIC, weight_decay=1e-5)
cnn_scheduler = optim.lr_scheduler.ReduceLROnPlateau(cnn_optimizer, 'min', patience=5, factor=0.2, verbose=True)
cnn_train_loss, cnn_val_loss, cnn_train_acc, cnn_val_acc = train_model(
    cnn_model, "CNN_Custom", train_loader_basic, val_loader_basic,
    common_criterion, cnn_optimizer, cnn_scheduler, epochs=EPOCHS_BASIC)
cnn_true_labels, cnn_preds, cnn_test_acc = test_model(
    cnn_model, "CNN_Custom", test_loader, common_criterion, classes)
model_results["CNN_Custom"] = {
    "train_loss_hist": cnn_train_loss, "val_loss_hist": cnn_val_loss, "train_acc_hist": cnn_train_acc, "val_acc_hist": cnn_val_acc,
    "test_labels": cnn_true_labels, "test_preds": cnn_preds, "test_acc": cnn_test_acc, "model_obj": cnn_model}
if cnn_train_loss: plot_learning_curves(cnn_train_loss, cnn_val_loss, cnn_train_acc, cnn_val_acc, "Custom CNN")
if cnn_true_labels: plot_confusion_matrix(cnn_true_labels, cnn_preds, classes, "Custom CNN")

print("\n Training ResNet18 ")
resnet_optimizer = optim.AdamW(resnet_model.parameters(), lr=LR_RESNET_MAX, weight_decay=WD_RESNET)
resnet_scheduler = optim.lr_scheduler.OneCycleLR(
    resnet_optimizer, max_lr=LR_RESNET_MAX, epochs=EPOCHS_RESNET,
    steps_per_epoch=len(train_loader_resnet), pct_start=0.3, div_factor=25, final_div_factor=1e4)
resnet_train_loss, resnet_val_loss, resnet_train_acc, resnet_val_acc = train_model(
    resnet_model, "ResNet18_Optimized", train_loader_resnet, val_loader_resnet,
    common_criterion, resnet_optimizer, resnet_scheduler, epochs=EPOCHS_RESNET, one_cycle_lr=True)
resnet_true_labels, resnet_preds, resnet_test_acc = test_model(
    resnet_model, "ResNet18_Optimized", test_loader, common_criterion, classes)
model_results["ResNet18_Optimized"] = {
    "train_loss_hist": resnet_train_loss, "val_loss_hist": resnet_val_loss, "train_acc_hist": resnet_train_acc, "val_acc_hist": resnet_val_acc,
    "test_labels": resnet_true_labels, "test_preds": resnet_preds, "test_acc": resnet_test_acc, "model_obj": resnet_model}
if resnet_train_loss: plot_learning_curves(resnet_train_loss, resnet_val_loss, resnet_train_acc, resnet_val_acc, "Optimized ResNet18")
if resnet_true_labels: plot_confusion_matrix(resnet_true_labels, resnet_preds, classes, "Optimized ResNet18")

print("\n\n FINAL TEST ACCURACY COMPARISON ")
print(f"{'Model':<25} | {'Test Accuracy':<20}")
print("-" * 50)
def display_test_acc(model_key, display_name_str):
    if model_key in model_results and model_results[model_key].get("test_acc") is not None:
        print(f"{display_name_str:<25} | {model_results[model_key]['test_acc']:.4f}")
    else:
        print(f"{display_name_str:<25} | {'N/A (Not trained or error)'}")
display_test_acc("MLP", "MLP")
display_test_acc("CNN_Custom", "Custom CNN (3 Conv)")
display_test_acc("ResNet18_Optimized", "Optimized ResNet18")
print("-" * 50)

print("\n VISUALIZING SAMPLE PREDICTIONS & PROBABILITIES ")
models_for_viz = {
    "MLP": model_results.get("MLP", {}).get("model_obj"),
    "CNN": model_results.get("CNN_Custom", {}).get("model_obj"),
    "ResNet18": model_results.get("ResNet18_Optimized", {}).get("model_obj")
}
active_models_for_viz = {k: v for k, v in models_for_viz.items() if v is not None}

if active_models_for_viz:
    visualize_predictions_with_probs(
        model_dict_param=active_models_for_viz,
        data_loader_param=test_loader,
        num_imgs=3,
        class_names_list=classes
    )
else:
    print("No models were fully trained to visualize predictions.")

```
